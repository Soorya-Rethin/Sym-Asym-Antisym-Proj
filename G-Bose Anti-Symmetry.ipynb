```python
def M(p, q):
    '''
    Unitary used to realize A(j).
    '''
    theta = 2*np.arctan(np.sqrt(q/p))
    qc = QuantumCircuit(1)
    qc.ry(theta, [0])
    return qc.to_gate()
```

```python
def cM(p, q):
    '''
    Controlled version of M(p, q).
    '''
    return M(p, q).control()
```

```python
def A(j):
    '''
    Unitary that prepares the state given in Equation~(12) from the paper.
    The (j-1)-qubit state is an equal superposition of the all-zeros state and 
    the set of computational basis state with a one in the ith position and zeros everywhere else.
    '''
    qc = QuantumCircuit(j-1)
    qc.append(M(1, j-1), [0])
    for i in range(j-2):
        qc.append(cM(1, j-2-i), [i, i+1])
    for i in range(j-2):
        qc.cx(i+1, i)
    return qc.to_gate()
```

```python
def createQuantumCircuit(state):
    '''
    Creates quantum circuit to test for anti-symmetry of a k-qubit input state.
    Refer to Section 2B.
    '''
    k = state.num_qubits
    
    #Create a quantum circuit with k*(k-1)/2 control qubits and k qubits for the input state.
    circ = QuantumCircuit(k*(k+1)/2, k*(k-1)/2)
    
    #Add the preparation unitaries. 
    startPos = 0
    for i in range(k-1):
        circ.append(A(k-i), range(startPos, startPos + k-i-1))
        startPos += (k-i-1)
    
    #Append the state preparation unitaries.
    circ.append(state, range(startPos, startPos+k))
    startState = startPos-1
    control_qubit = startPos-1
    
    #Add the controlled SWAP gates.
    for j in range(2, k+1):
        for i in range(j-1):
            circ.cswap(control_qubit, startState+j-i-1, startState+j)
            control_qubit -= 1
    
    circ.barrier()
    
    #Add the Z gates.
    circ.z(range(int(k*(k-1)/2)))
    circ.barrier()
    
    #Add the inverse of the preparation unitaries.
    startPos = 0
    for i in range(k-1):
        circ.append(A(k-i).inverse(), range(startPos, startPos + k-i-1))
        circ.measure(range(startPos, startPos + k-i-1), range(startPos, startPos + k-i-1))
        startPos += (k-i-1)
    
    return circ
```

```python
state = QuantumCircuit(3)
state.h(0)
state.cx(0, 1)
state.x(1)
state.z(1)
state.draw('mpl')
```

```python
qCirc = createQuantumCircuit(state)
qCirc.draw('mpl')
```

```python
sampler = prim_Sampler()
dic = sampler.run(qCirc).result().quasi_dists[0]
print(dic)
```
