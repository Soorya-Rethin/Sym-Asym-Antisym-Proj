```python
def get_from_file(name):
    file = open(name, "r")
    string = file.read()
    file.close()
    lis = string.split("\n")
    qasm_reconstructed = ""
    for i in range(len(lis)):
        if(lis[i][0] == "{"):
            break
        qasm_reconstructed += lis[i]
        qasm_reconstructed += "\n"

    lis_exp = [eval(lis[j]) for j in list(range(i, i+4))]
    
    quantum_backend_name = lis[-2]
    
    return qasm_reconstructed, lis_exp, quantum_backend_name
```

```python
noiseless = {}
shot_noise = {}
noisy = {}
quantum = {}
```

```python
r = 2
filename = "Schmidt Rank Data Files/3-qubit-W-2-1-"+str(r)
outname = "Schmidt Rank Data Files/3-qubit-W-2-1"
qasm_reconstructed, lis_exp, quantum_name = get_from_file(filename+".txt")
```

```python
if 0 in lis_exp[0]:
    noiseless[str(r)] = lis_exp[0][0]
else:
    noiseless[str(r)] = 0.0
```

```python
if 0 in lis_exp[1]:
    shot_noise[str(r)] = lis_exp[1][0]
else:
    shot_noise[str(r)] = 0.0
```

```python
if 0 in lis_exp[2]:
    noisy[str(r)] = lis_exp[2][0]
else:
    noisy[str(r)] = 0.0
```

```python
if 0 in lis_exp[3]:
    quantum[str(r)] = lis_exp[3][0]
else:
    quantum[str(r)] = 0.0
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a new figure
fig = plt.figure(figsize = (15, 10))

# Add a subplot to the figure
ax = fig.add_subplot(111)

# Extract keys and values for each dictionary
labels = list(noiseless.keys())
values1 = list(noiseless.values())
values2 = list(shot_noise.values())
values3 = list(noisy.values())
values4 = list(quantum.values())

# Determine the number of dictionaries and the width of each bar
n_dicts = 4
bar_width = 0.2
x = np.arange(len(labels))

# Plot each dictionary's bars with an offset to avoid overlap
ax.bar(x - bar_width, values1, width=bar_width, color='black', label='Noiseless')
ax.bar(x, values2, width=bar_width, color='gray', label='Shot Noise')
ax.bar(x + bar_width, values3, width=bar_width, color='silver', label='Noisy Simulation')
ax.bar(x + 2*bar_width, values4, width=bar_width, color='lightgray', label='Quantum Device ' + quantum_name)

# Add titles and labels
ax.set_ylabel("Probability of outcome 0", **timesfont,fontsize=35)
ax.set_xlabel(r"Schmidt Rank Cutoff $\it{r}$", **timesfont,fontsize=35)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.tick_params(axis='x', labelsize=25)
ax.tick_params(axis='y', labelsize=25)

# Add a legend
ax.legend(loc = 0, fontsize=30)
ax.grid(axis='y')

# Show the plot
plt.show()
#fig.savefig(outname+"BW.pdf", dpi=2000, format='pdf', bbox_extra_artists=(), bbox_inches='tight')
```
